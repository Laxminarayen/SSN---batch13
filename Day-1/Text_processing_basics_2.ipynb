```python
from nltk.corpus import gutenberg
nltk.download('gutenberg')
```

```python
import seaborn as sns
%matplotlib inline
```

```python
bible = gutenberg.open('bible-kjv.txt')
```

```python
bible = bible.readlines()
```

```python
bible[:5]
```

```python
len(bible)
```

```python
[item.strip('\n') for item in bible[:5]]
```

```python
bible = list(filter(None, [item.strip('\n') for item in bible]))
bible[:5]
```

```python
line_legths = [len(sentence) for sentence in bible]
p = sns.kdeplot(line_legths,shade = True,color = 'red')
```

## Finding word count distribution

```python
tokens = [item.split() for item in bible]
print(tokens[:5])
```

```python
line_legths = [len(sentence) for sentence in tokens]
p = sns.kdeplot(line_legths,shade = True,color = 'red')
```

#### SYNTAX: 
[ the append statement , First for loop, Second for loop]

```python
words = [item for word in tokens for item in word]
```

```python
import re
list(filter(None,[re.sub(r'[^A-Za-z]','',word) for word in words]))
```
